```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Resources/lena.png")
```

```python
kernel = np.ones((5, 5), np.uint8)
```

### Basic Processing of Images:

```python
# Gray-scale image:
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Blurred image:
imgBlur = cv2.GaussianBlur(imgGray, (7, 7), 0)  # Parameters: grayImage, kernelSize, sigmaSize
```

```python
# Canny image:
imgCanny = cv2.Canny(img, 100, 100)  # Parameters: img, threshold1, threshold2
```

```python
# Dilated image:
imgDilation = cv2.dilate(imgCanny, kernel, iterations=1)
```

```python
# Eroded image:
imgEroded = cv2.erode(imgDilation, kernel, iterations=1)
```
